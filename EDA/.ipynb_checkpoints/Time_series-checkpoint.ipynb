```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact
import seaborn as sns
```

```python
df = pd.read_csv('Go_daddy_combined_data\Go_daddy_combined_data.csv')
df
```

```python
df['first_day_of_month_orig'] = pd.to_datetime(df['first_day_of_month_orig'])
```

```python
df.dtypes
```

```python
drop_down = widgets.Dropdown(options=df.county.unique(), description='Select a county')
drop_down
```

```python
def draw_lineplot(county):
    data = df.query(f"county == '{county}'")
    fig = sns.lineplot(data=data, x="first_day_of_month_orig", y="microbusiness_density")
    #fig.write_html('first_figure.html', auto_open=True)
```

```python
interact(draw_lineplot, county=drop_down)
```
